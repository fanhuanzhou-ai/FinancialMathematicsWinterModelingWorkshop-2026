```python
import pandas as pd
import numpy as np
import scipy.stats as ss
import statsmodels.api as sm
```

First I am inputting the Data from the Portfolio tabs I created from the excel workshop data, and the new Indexes/YAS files obtained.

```python
portfolio = pd.read_excel(io="Workshop Data.xlsx",sheet_name="Portfolio")
portfolio = portfolio.set_index("Date")
portfolio.head()
```

```python
HYGVol = pd.read_excel(io="Workshop Data.xlsx",sheet_name="Adj HYG")
HYGVol = HYGVol.set_index("Date")
HYGVol = HYGVol.loc[:,"Volume"]
HYGVol.head()
```

```python
LQDyas = pd.read_excel(io="Indexes and Spreads Data 01.09.xlsx",sheet_name="LQD")
LQDyas = LQDyas.set_index("Date")
LQDyas.head()
```

```python
HYGyas = pd.read_excel(io="Indexes and Spreads Data 01.09.xlsx",sheet_name="HYG")
HYGyas = HYGyas.set_index("Date")
HYGyas = pd.DataFrame(HYGyas)
HYGyas.head()
```

```python
TenYearYield = pd.read_excel(io="Indexes and Spreads Data 01.09.xlsx",sheet_name="10yUST Yields")
TenYearYield = TenYearYield.set_index("Date")
TenYearYield.head()
```

```python
HYyield = pd.read_excel(io="Indexes and Spreads Data 01.09.xlsx",sheet_name="HY Index")
HYyield = HYyield.set_index("Date")
HYyield.head()
```

```python
IGyield = pd.read_excel(io="Indexes and Spreads Data 01.09.xlsx",sheet_name="IG Index")
IGyield = IGyield.set_index("Date")
IGyield.head()
```

```python
portfolio.loc[:,"HYGCumDiv"]=portfolio.loc[:,"HYG Dividends"][::-1].cumsum()[::-1]
portfolio.tail()
```

```python
portfolio.loc[:,"TotalReturnsHYG"]=portfolio.loc[:,"HYG Position"]+portfolio.loc[:,"HYGCumDiv"]
portfolio.tail()
```

Now, I am limiting the dates index so that all the data have the same number of entries and is consistent.

```python
portfolio = portfolio.sort_index()
HYyield = HYyield.sort_index()
IGyield = IGyield.sort_index()
TenYearYield = TenYearYield.sort_index()
HYGVol = HYGVol.sort_index()
HYGyas = HYGyas.sort_index()
HYGCorrData = pd.DataFrame(portfolio.loc["2021-01-11":"2025-12-31","TotalReturnsHYG"]).merge(HYyield.loc["2021-01-11":"2025-12-31"],left_index=True,right_index=True)
HYGCorrData = HYGCorrData.rename(columns={"PX_LAST":"HY_INDEX_PX","OAS_SOVEREIGN_CURVE":"HY_Index_OAS","INDEX_SPREAD_BENCHMARK":"HY_INDEX_Spread","INDEX_Z_SPREAD_BP":"HY_INDEX_Z_Spread"})
HYGCorrData = HYGCorrData.loc["2021-01-11":"2025-12-31"].merge(TenYearYield.loc["2021-01-11":"2025-12-31","PX_LAST"],left_index=True,right_index=True)
HYGCorrData = HYGCorrData.rename(columns={"PX_LAST":"10y_UST_INDEX_PX"})
HYGCorrData = HYGCorrData.loc["2021-01-11":"2025-12-31"].merge(IGyield.loc["2021-01-11":"2025-12-31"],left_index=True,right_index=True)
HYGCorrData = HYGCorrData.rename(columns={"PX_LAST":"IG_INDEX_PX","OAS_SOVEREIGN_CURVE":"IG_Index_OAS","INDEX_SPREAD_BENCHMARK":"IG_INDEX_Spread","INDEX_Z_SPREAD_BP":"IG_INDEX_Z_Spread"})
HYGCorrData = HYGCorrData.loc["2021-01-11":"2025-12-31"].merge(HYGVol.loc["2021-01-11":"2025-12-31"],left_index=True,right_index=True)
HYGCorrData = HYGCorrData.rename(columns={"Volume":"HYG_Volume"})
HYGCorrData = HYGCorrData.loc["2021-01-11":"2025-12-31"].merge(portfolio.loc["2021-01-11":"2025-12-31","SPY Position"]/-10,left_index=True,right_index=True)
HYGCorrData = HYGCorrData.rename(columns={"SPY Position":"SPY_LAST_PX"})
HYGCorrData = HYGCorrData.loc["2021-01-11":"2025-12-31"].merge(HYGyas.loc["2021-01-11":"2025-12-31","YAS_MOD_DUR"],left_index=True,right_index=True)
HYGCorrData = HYGCorrData.rename(columns={"YAS_MOD_DUR":"HYG_MOD_DUR_YAS"})

HYGCorrData = HYGCorrData.sort_index(ascending=False)


HYGCorrData.head()
```

**Correlation Table**

```python
HYGCorr = HYGCorrData.corr()
HYGCorr
```

Following, I created lagged variables by one day of all the previous information to run the regressions.

```python
HYGCorrDataLagVol = HYGCorrData.copy()
HYGCorrDataLagVol.loc[:,"HY_Index_OAS_Lag_1"]=HYGCorrDataLagVol.loc[:,"HY_Index_OAS"].shift(-1)
HYGCorrDataLagVol.loc[:,"HY_INDEX_PX_Lag_1"]=HYGCorrDataLagVol.loc[:,"HY_INDEX_PX"].shift(-1)
HYGCorrDataLagVol.loc[:,"HY_INDEX_Spread_Lag_1"]=HYGCorrDataLagVol.loc[:,"HY_INDEX_Spread"].shift(-1)
HYGCorrDataLagVol.loc[:,"HY_INDEX_Z_Spread_Lag_1"]=HYGCorrDataLagVol.loc[:,"HY_INDEX_Z_Spread"].shift(-1)
HYGCorrDataLagVol.loc[:,"10y_UST_INDEX_PX_Lag_1"]=HYGCorrDataLagVol.loc[:,"10y_UST_INDEX_PX"].shift(-1)
HYGCorrDataLagVol.loc[:,"IG_Index_OAS_Lag_1"]=HYGCorrDataLagVol.loc[:,"IG_Index_OAS"].shift(-1)
HYGCorrDataLagVol.loc[:,"IG_Index_Spread_Lag_1"]=HYGCorrDataLagVol.loc[:,"IG_INDEX_Spread"].shift(-1)
HYGCorrDataLagVol.loc[:,"IG_INDEX_PX_Lag_1"]=HYGCorrDataLagVol.loc[:,"IG_INDEX_PX"].shift(-1)
HYGCorrDataLagVol.loc[:,"IG_INDEX_Z_Spread_Lag_1"]=HYGCorrDataLagVol.loc[:,"IG_INDEX_Z_Spread"].shift(-1)
HYGCorrDataLagVol.loc[:,"HYG_Volume_Lag_1"]=HYGCorrDataLagVol.loc[:,"HYG_Volume"].shift(-1)
HYGCorrDataLagVol.loc[:,"SPY_LAST_PX_Lag_1"]=HYGCorrDataLagVol.loc[:,"SPY_LAST_PX"].shift(-1)
HYGCorrDataLagVol.loc[:,"HYG_MOD_DUR_YAS_Lag_1"]=HYGCorrDataLagVol.loc[:,"HYG_MOD_DUR_YAS"].shift(-1)
HYGCorrDataLagVol.loc[:,"HY_OASxIG_OAS_Lag_1"]=(HYGCorrDataLagVol.loc[:,"HY_Index_OAS_Lag_1"]*HYGCorrDataLagVol.loc[:,"IG_Index_OAS_Lag_1"])
HYGCorrDataLagVol=HYGCorrDataLagVol.dropna()
HYGCorrDataLagVol.tail()
```

I tried several OLS regressions and all of them have a high multicolineariy indicatior of conditional number. For example look at the one below:

```python
X = sm.add_constant(HYGCorrDataLagVol[["HY_Index_OAS","10y_UST_INDEX_PX","IG_Index_OAS","SPY_LAST_PX","HY_OASxIG_OAS"]])
Y = HYGCorrDataLagVol["TotalReturnsHYG"]/100

model = sm.OLS(Y,X).fit()
model.summary()
```

For this reason, I am limited the OLS time series regressions to the Option adjusted spreads from UST for both High Yields and Investment grades bonds (including a combined dependent variable), and the 10 Year US Treasury yields. 

My logic says that the price and returns would depende on how much is paying over the UST, and if the Spread is high for investment grades & lower than usual for high yield then people would switch from one asset class to the other, which would reduce the price by demand. 

```python
X = sm.add_constant(HYGCorrDataLagVol[["HY_Index_OAS","10y_UST_INDEX_PX","IG_Index_OAS","HY_OASxIG_OAS"]])
Y = HYGCorrDataLagVol["TotalReturnsHYG"]/100

model = sm.OLS(Y,X).fit()
model.summary()
```

Let's Try above expression lagged by one Day to see if that solves the structure


```python
X = sm.add_constant(HYGCorrDataLagVol[["HY_Index_OAS_Lag_1","10y_UST_INDEX_PX_Lag_1","IG_Index_OAS_Lag_1","HY_OASxIG_OAS_Lag_1"]])
Y = HYGCorrDataLagVol["TotalReturnsHYG"]/100

model = sm.OLS(Y,X).fit()
model.summary()
```

The conditional number is still pretty high: 247, I decided to examine each variable one by one, and it increases really fast.

```python
X = sm.add_constant(HYGCorrDataLagVol[["10y_UST_INDEX_PX"]])
Y = HYGCorrDataLagVol["TotalReturnsHYG"]/100

model = sm.OLS(Y,X).fit()
model.summary()
```

```python
X = sm.add_constant(HYGCorrDataLagVol[["10y_UST_INDEX_PX","HY_Index_OAS"]])
Y = HYGCorrDataLagVol["TotalReturnsHYG"]/100

model = sm.OLS(Y,X).fit()
model.summary()
```

```python
X = sm.add_constant(HYGCorrDataLagVol[["10y_UST_INDEX_PX","HY_Index_OAS","HYG_MOD_DUR_YAS"]])
Y = HYGCorrDataLagVol["TotalReturnsHYG"]/100

model = sm.OLS(Y,X).fit()
model.summary()
```

```python
X = sm.add_constant(HYGCorrDataLagVol[["10y_UST_INDEX_PX_Lag_1","HY_Index_OAS_Lag_1","HYG_MOD_DUR_YAS_Lag_1"]])
Y = HYGCorrDataLagVol["TotalReturnsHYG"]/100

model = sm.OLS(Y,X).fit()
model.summary()
```

**Final regression I obtained so far**

Has a 288 conditional number, p-values that can not deny the null hypothesis and a R-Squared that imply the variables could explain an 81% of the dependent variable variance. Do you think the Conditional Number is a concern? Once I add SPY variables the conditional number goes from a little high to extremely concerning. 

```python
X = sm.add_constant(HYGCorrDataLagVol[["10y_UST_INDEX_PX_Lag_1","HY_Index_OAS_Lag_1","IG_Index_OAS_Lag_1","HY_OASxIG_OAS_Lag_1","HYG_MOD_DUR_YAS_Lag_1"]])
Y = HYGCorrDataLagVol["TotalReturnsHYG"]/100

model = sm.OLS(Y,X).fit()
model.summary()
```
